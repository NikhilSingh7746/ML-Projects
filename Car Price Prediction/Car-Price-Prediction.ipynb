```python
# Import Dependencies
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from warnings import filterwarnings
filterwarnings('ignore')

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression , Lasso
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Import DS
data = pd.read_csv('/content/drive/MyDrive/ML-Projects/car data.csv')
data.head()
```

```python
data.info()
```

Data Preprocessing

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(keep='last',inplace=True)
```

```python
print(data['Fuel_Type'].value_counts())
print(data['Seller_Type'].value_counts())
print(data['Transmission'].value_counts())
```

```python
# Encoding
data['Fuel_Type'] = data['Fuel_Type'].map({'Petrol':0,'Diesel':1,'CNG':2})
data['Seller_Type'] = data['Seller_Type'].map({'Dealer':0,'Individual':1})
data['Transmission'] = data['Transmission'].map({'Manual':0,'Automatic':1})
```

```python
# Define Ip & Op
x = data.drop(columns=['Car_Name','Selling_Price'])
y = data['Selling_Price']
```

```python
# Split the DS
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

Model Training :- Linear Regression

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred_train = lr.predict(x_train)
y_pred_test = lr.predict(x_test)
```

```python
r2 = r2_score(y_test,y_pred_test)
print('R2 Score : ',r2)
```

```python
plt.scatter(y_train, y_pred_train)
plt.scatter(y_test, y_pred_test, color='red')
plt.show()
```

Model Training - Lasso Regression

```python
lasso = Lasso()
lasso.fit(x_train,y_train)
```

```python
l_y_pred_train = lasso.predict(x_train)
l_y_pred_test = lasso.predict(x_test)
```

```python
r = r2_score(y_test,l_y_pred_test)
print('R2 Score : ',r)
```
